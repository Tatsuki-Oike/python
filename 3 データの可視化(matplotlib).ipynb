# 1 ライブラリ

```python
import numpy as np
import matplotlib.pyplot as plt
```

# 2 データの可視化

```python
x = np.arange(10)
y1 = 2*x+3
plt.plot(x,y1)
```

```python
np.random.seed(1)
x = np.arange(10)
y2 = 2*x+3+np.random.randn(10)
plt.scatter(x,y2)
```

# 3 色々な可視化

## 3.1 ラベル追加

```python
plt.plot(x,y1)
plt.xlabel("x")
plt.ylabel("y")
plt.title("title")
```

## 3.2 複数同時プロット

```python
plt.plot(x,y1)
plt.scatter(x,y2)
```

```python
plt.plot(x,y1,label="label1")
plt.scatter(x,y2,label="label2")
plt.legend()
```

```python
figure = plt.figure(figsize=(15,8))
ax = figure.add_subplot(2,2,1)
ax.plot(x,y1)
ax = figure.add_subplot(2,2,2)
ax.plot(x,y1)
ax = figure.add_subplot(2,2,3)
ax.scatter(x,y2)
ax = figure.add_subplot(2,2,4)
ax.scatter(x,y2)
```
